```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
import plotly
```

# Load the Data

```python
# Load data
df = pd.read_csv('/content/Sample - Superstore.csv', encoding='latin1')

# Quick view
print(df.head())
```

```python
df.info()
```

```python
df.head(5)
```

```python
df.tail(5)
```

# Data Cleaning

```python
# Check for nulls
print(df.isnull().sum())
```

```python
# I will dropped some un-necessary columns
print(df.columns)  # List all column names
```

```python
df_1 = df.drop(['Row ID', 'Postal Code','Segment','Customer Name','Ship Mode','Country','Product Name'], axis=1)
print(df_1.columns)
```

```python
df_1.head(5)
```

```python
# Convert Order Date to datetime
df_1['Order Date'] = pd.to_datetime(df['Order Date'])
```

# Descriptive Analysis


```python
df_1.describe()
```

 # Total Sales & Profit

```python
total_sales = df_1['Sales'].sum()
total_profit = df_1['Profit'].sum()

print(f"Total Sales: ₹{total_sales:,.2f}")
print(f"Total Profit: ₹{total_profit:,.2f}")
```

# Top 10 Products by Sales

```python
top_products = df.groupby('Sub-Category')['Sales'].sum().sort_values(ascending=False).head(10)
print(top_products)
```

# 📈  Trend Analysis

**🔸 Monthly Sales Trend**

```python
# prompt: 🔸 Monthly Sales Trend

# Set 'Order Date' as index
df_1.set_index('Order Date', inplace=True)

# Resample data by month and sum sales
monthly_sales = df_1['Sales'].resample('M').sum()

# Plot monthly sales trend
plt.figure(figsize=(12, 6))
monthly_sales.plot()
plt.title('Monthly Sales Trend')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.grid(True)
plt.show()
```

# 🔸 Sales by Category & Sub-Category

```python
plt.figure(figsize=(10,6))
sns.barplot(data=df, x='Category', y='Sales', estimator=sum, ci=None)
plt.title("Total Sales by Category")
plt.show()
```

# 🔸 Top 5 Cities by Sales

```python
top_cities = df.groupby('City')['Sales'].sum().sort_values(ascending=False).head(5)
print(top_cities)
```

# 📉 Step 8: Profit Loss Analysis
🔹 Sub-categories with Loss

```python
loss_data = df.groupby('Sub-Category')['Profit'].sum().sort_values()
print(loss_data[loss_data < 0])
```
